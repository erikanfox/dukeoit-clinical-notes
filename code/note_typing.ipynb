```python
# create feature 'type' that condenses 'title','admission_service' and 'specialty'
notes["type"]=np.nan
notes["type"]=np.where(notes["title"].notnull(),notes["title"],notes["type"])
notes["type"]=np.where(notes["admission_service"].notnull() & notes["type"].isnull(),notes["admission_service"],notes["type"])
notes["type"]=np.where(notes["specialty"].notnull() & notes["type"].isnull(),notes["specialty"],notes["type"])
```

```python
# edit values of type, so every single specialty has just one name

notes["type"]=notes["type"].str.upper()
notes["type"]=np.where(notes["type"]=="HEMATOLOGY","HEMATOLOGY AND ONCOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="INFECTIOUS DISEASE","INFECTIOUS DISEASES",notes["type"])
notes["type"]=np.where(notes["type"]=="GYN ONCOLOGY","GYNECOLOGIC ONCOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="ONCOLOGY (DEPT OF MED)","ONCOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PHO - PEDIATRIC HEMATOLOGY AND ONCOLOGY","PEDIATRIC HEMATOLOGY AND ONCOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PED HEMATOL-ONCOLOGY","PEDIATRIC HEMATOLOGY AND ONCOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PEDIATRIC ALLERGY IMMUNOLOGY","PEDIATRIC ALLERGY AND IMMUNOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PED ALLER-IMMUNO-PUL","PEDIATRIC ALLERGY AND IMMUNOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PED ENDOCRINOLOGY","PEDIATRIC ENDOCRINOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PED GASTROENTEROLOGY","PEDIATRIC GASTROENTEROLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="MATERNAL AND FETAL MEDICINE","OBGYN",notes["type"])
notes["type"]=np.where(notes["type"]=="LABOR AND DELIVERY","OBGYN",notes["type"])
notes["type"]=np.where(notes["type"]=="UROGYNECOLOGY","OBGYN",notes["type"])
notes["type"]=np.where(notes["type"]=="GYNECOLOGY","OBGYN",notes["type"])
notes["type"]=np.where(notes["type"]=="OBSTETRICS","OBGYN",notes["type"])
notes["type"]=np.where(notes["type"]=="OBSTETRICS AND GYNECOLOGY","OBGYN",notes["type"])
notes["type"]=np.where(notes["type"]=="REPRODUCTIVE ENDOCRINOLOGY AND INFERTILITY","OBGYN",notes["type"])
notes["type"]=np.where(notes["type"]=="PEDIATRIC PULMONARY","PEDIATRIC PULMONOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PEDIATRIC PRIVATE","PEDIATRICS",notes["type"])
notes["type"]=np.where(notes["type"]=="GENERAL PEDIATRICS","PEDIATRICS",notes["type"])
notes["type"]=np.where(notes["type"]=="PED NEONATOLOGY","PEDIATRIC NEONATOLOGY AND PERINATOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="NEONATAL-PERINATAL PEDIATRICS","PEDIATRIC NEONATOLOGY AND PERINATOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="PERINATOLOGY","PEDIATRIC NEONATOLOGY AND PERINATOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="FAMILY MEDICINE","PRIMARY CARE",notes["type"])
notes["type"]=np.where(notes["type"]=="GENERAL INTERNAL MEDICINE","PRIMARY CARE",notes["type"])
notes["type"]=np.where(notes["type"]=="INTERNAL MEDICINE","PRIMARY CARE",notes["type"])
notes["type"]=np.where(notes["type"]=="OUTPATIENT","PRIMARY CARE",notes["type"])
notes["type"]=np.where(notes["type"]=="URGENT CARE","PRIMARY CARE",notes["type"])
notes["type"]=np.where(notes["type"]=="EAR NOSE THROAT","OTOLARYNGOLOGY",notes["type"])
notes["type"]=np.where(notes["type"]=="TRAUMA","TRAUMA SURGERY",notes["type"])
notes["type"]=np.where(notes["type"]=="UROLOGY","UROLOGIC SURGERY",notes["type"])
notes["type"]=np.where(notes["type"]=="HOSPITAL MEDICINE","INTENSIVE CARE",notes["type"])
```
